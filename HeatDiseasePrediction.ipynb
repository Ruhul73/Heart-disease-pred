## Hyperparameter Tuning and Cross-Validation for Random Forest

### Subtask:
Perform hyperparameter tuning for the Random Forest Classifier using GridSearchCV or RandomizedSearchCV with cross-validation. This will help optimize parameters like 'n_estimators', 'max_depth', and 'min_samples_leaf' to prevent overfitting and improve the model's predictive power on unseen data.


**Reasoning**:

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_leaf': [1, 2, 4]
}

rfc = RandomForestClassifier(random_state=42)

grid_search_rf = GridSearchCV(rfc, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

grid_search_rf.fit(X_train_scaled, y_train)

print("Best parameters found for Random Forest: ", grid_search_rf.best_params_)
print("Best accuracy score for Random Forest: ", grid_search_rf.best_score_)
````

## ROC Curve and AUC for Best Models

### Subtask:
Generate ROC (Receiver Operating Characteristic) curves and calculate the AUC (Area Under the Curve) score for the top-performing models (tuned KNN and tuned Random Forest).


**Reasoning**:
As per the instructions, I will now import the necessary modules, retrieve the best estimators for KNN and Random Forest, predict probabilities on the scaled test data, calculate FPR, TPR, and AUC for both models, and finally plot their ROC curves along with a random classifier baseline. This will complete the subtask of generating ROC curves and AUC scores for the top-performing models.



````python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# 2. Retrieve the best K-Nearest Neighbors estimator
best_knn_model = grid_search.best_estimator_

# 3. Predict probabilities for the positive class (y=1) on X_test_scaled using the best KNN estimator
y_pred_proba_knn = best_knn_model.predict_proba(X_test_scaled)[:, 1]

# 4. Calculate FPR, TPR, and AUC for the KNN model
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, y_pred_proba_knn)
auc_knn = auc(fpr_knn, tpr_knn)

# 5. Retrieve the best Random Forest estimator
best_rf_model = grid_search_rf.best_estimator_

# 6. Predict probabilities for the positive class (y=1) on X_test_scaled using the best Random Forest estimator
y_pred_proba_rf = best_rf_model.predict_proba(X_test_scaled)[:, 1]

# 7. Calculate FPR, TPR, and AUC for the Random Forest model
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_proba_rf)
auc_rf = auc(fpr_rf, tpr_rf)

# 8. Create a new figure and axes for the plot
plt.figure(figsize=(10, 8))

# 9. Plot the ROC curve for the tuned KNN model
plt.plot(fpr_knn, tpr_knn, color='blue', lw=2, label='Tuned KNN (AUC = %0.2f)' % auc_knn)

# 10. Plot the ROC curve for the tuned Random Forest model
plt.plot(fpr_rf, tpr_rf, color='red', lw=2, label='Tuned Random Forest (AUC = %0.2f)' % auc_rf)

# 11. Plot a diagonal dashed line representing a random classifier
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--', label='Random Classifier')

# 12. Set the title of the plot
plt.title('ROC Curves for Tuned KNN and Random Forest')

# 13. Label the x-axis and y-axis
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# 14. Add a legend to the plot
plt.legend(loc='lower right')

# 15. Display the plot
plt.show()

print(f"Tuned KNN AUC: {auc_knn:.2f}")
print(f"Tuned Random Forest AUC: {auc_rf:.2f}")
````

## Analyze Perfect Scores

### Subtask:
Explain that perfect scores are highly unusual and often indicate issues such as overfitting or data leakage. Specifically address why KNN with n_neighbors=1 might show perfect scores due to memorizing the training data. Also, mention that Decision Trees can also easily overfit.


### Analyzing Perfect Scores

Perfect scores (e.g., 1.0 for accuracy or AUC) on a model are highly unusual and often indicative of underlying issues rather than true perfect generalization. In real-world datasets, achieving such scores typically points to problems like **overfitting** or **data leakage**.

#### Overfitting
Overfitting occurs when a model learns the training data, including its noise and outliers, too well. This results in excellent performance on the training set but poor generalization to new, unseen data.

*   **K-Nearest Neighbors (KNN) with `n_neighbors=1`**: When `n_neighbors` is set to 1, the KNN model simply classifies a new data point based on its closest neighbor in the training set. If the test set contains samples that are identical or very similar to those in the training set (e.g., due to accidental data duplication or if the test set is not truly independent), a `n_neighbors=1` KNN model will essentially 'memorize' the training data. This memorization leads to perfect predictions on those specific test instances, artificially inflating the score.

*   **Decision Trees**: Unconstrained Decision Tree models are also highly prone to overfitting. Without limits on depth (`max_depth`) or minimum samples per leaf (`min_samples_leaf`), a Decision Tree can create very complex and specific rules that perfectly fit every data point in the training set. This high complexity allows it to achieve 100% accuracy on the training data, but it loses its ability to generalize to slightly different new data points, leading to poor performance on unseen data. The exact structure of the tree becomes too tailored to the training set's quirks.

### Summary of Perfect Scores and Recommendations

The previous data integrity checks have confirmed that the `X_train`, `X_test`, `y_train`, and `y_test` splits do not have overlapping indices and their combined lengths match the original dataset. This indicates that the perfect scores observed are unlikely due to accidental data leakage through the splitting process.

However, **perfect scores (1.0 AUC and accuracy)** are still highly suspicious in real-world datasets and strongly suggest **overfitting**. As previously discussed:

*   **K-Nearest Neighbors (KNN) with `n_neighbors=1`**: An `n_neighbors=1` setting makes the model highly susceptible to noise in the training data. If the test set contains points very close or identical to the training points (even without explicit data leakage via indices, due to the nature of the data itself or insufficient variability in small datasets), it can lead to perfect predictions. The previous elbow method showed varying accuracy, implying that `n_neighbors=1` was indeed picking up on specific, possibly noisy, patterns that don't generalize.

*   **Decision Trees**: Unconstrained Decision Trees can create a highly complex model that memorizes the training data perfectly, resulting in 100% accuracy on the training set and potentially overfitting to the test set as well if it's not truly diverse or sufficiently large. Even though hyperparameter tuning was performed for Random Forest (an ensemble of Decision Trees), the base Decision Tree itself showed perfect scores.

This technique is designed to provide a more robust estimate of model performance and help select optimal hyperparameters by evaluating different parameter combinations across multiple splits of the training data.

2.  **Evaluation on a Single Test Set**: Despite the use of cross-validation during tuning, the final 'perfect' scores (e.g., 1.0 accuracy or AUC) were reported based on the evaluation of the *best* tuned model on a single, fixed test set (`X_test`, `y_test`).

3.  **Limitations of Single Test Set Evaluation with Perfect Scores**: While cross-validation helps to find better parameters that generalize well on average across different training-validation splits, achieving perfect scores on a single, final test set warrants close inspection. A single test set might not be representative enough to fully capture the true generalization ability of a model, especially if its size is small or if it shares unexpected similarities with the training data. If the test set happens to be 'easy' for the model to predict, or if certain data characteristics lead to an over-optimistic performance, perfect scores can occur.

4.  **Robust Evaluation and Scrutiny**: Robust evaluation typically requires more than just performance on a single test set. Even when hyperparameter tuning with cross-validation has been performed, perfect scores should always be scrutinized. They can still signal issues like:
    *   **Overfitting to the test set**: While less common with proper split, it's possible if the test set is too similar to the training data or if there's an implicit data leakage during the model development process.
    *   **Data Imbalance**: If one class heavily dominates and the model simply predicts that class, it might achieve high accuracy if not evaluated with other metrics.
    *   **Feature Engineering Issues**: Features that inadvertently contain information about the target variable from the future or from the test set can lead to inflated scores.
    *   **Trivial Problem**: The problem itself might be too simple, though this is rare in real-world scenarios that necessitate machine learning.

Therefore, while the hyperparameter tuning process aimed to improve generalization, the consistent perfect scores across models on the test set require further investigation beyond standard metrics to ensure the models are truly robust and not just 'lucky' on this particular split.

This variability suggests that the choice of `n_neighbors` has a substantial impact on the model's performance, as expected. However, it's crucial to note the first entry in this list: a perfect score of `100` (representing 100% accuracy) when `n_neighbors=1`. Subsequently, the accuracy drops considerably for `n_neighbors=2` (93%), `n_neighbors=3` (89%), and even more for larger values of `k`.

**Contrast with Perfect Score for `n_neighbors=1` and Overfitting**:

The stark contrast between the perfect score for `n_neighbors=1` and the significantly lower (though still good) scores for other `k` values is a strong indicator of **overfitting**. When `n_neighbors=1`, the KNN model essentially 'memorizes' the training data. For every test data point, it simply finds the single closest training data point and assigns its class. If the test data contains instances that are very similar or even identical to the training data (which can happen, especially in smaller datasets or if the train-test split wasn't perfectly randomized), a `k=1` model will achieve perfect or near-perfect accuracy on the test set, but this performance is often misleading.

This behavior means the model is not generalizing well; instead, it is highly sensitive to the specific nuances and noise of the training data. While a perfect score might seem desirable, in machine learning, it almost always signals a problem with the model or the evaluation process rather than true, robust performance. It suggests that the model has learned the training data too well, to the detriment of its ability to predict unseen, slightly different data points accurately.
